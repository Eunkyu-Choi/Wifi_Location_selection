```python
wifi_out = wifi[wifi['유형']!='공공기관']
wifi_out = wifi_out[wifi_out['유형']!='복지센터등']
wifi_out = wifi_out[wifi_out['유형']!='문화관광']
wifi_out.shape
```

```python
wifi_out1 = gpd.GeoDataFrame(wifi_out, geometry = gpd.points_from_xy(wifi_out['설치위치(X좌표)'], wifi_out['설치위치(Y좌표)']))
```

```python
colnames = ['gu','type','address','name','detail_add','com','X','Y','ID','geometry']
wifi_out1.columns = colnames
```

```python
wifi_out1.to_file('wifi_out.shp', encoding = 'cp949')
```

```python
import json
import folium
```

```python
map = folium.Map(location=[wifi['설치위치(Y좌표)'].mean(), wifi['설치위치(X좌표)'].mean()], zoom_start=12, tiles='OpenStreetMap')
for n in wifi.index:
    popup_name = wifi['구명'][n]

    folium.CircleMarker(
        location=[wifi['설치위치(Y좌표)'][n], wifi['설치위치(X좌표)'][n]],
        radius=5,
        popup=popup_name,
        fill=True,
        color='#3186cc'
    ).add_to(map)
map
```
